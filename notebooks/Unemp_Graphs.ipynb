```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.graph_objs as go
import psutil
import plotly.io as pio
from Unemployment import *
from Suicide import *
```

```python
ucm = go.Figure()
ucm.add_trace(go.Bar(x=unemp_countries_mean.index, y=unemp_countries_mean.unemploy_rate,  name="Unemployment Mean Rate 200-2016"))
ucm.update_layout(legend=dict(x=0,y=1), width=1400, height=750, title='Unemployment Mean Rate Per Country % (2000 - 2016)')
degrees = 310
ucm.update_layout(xaxis=dict(tickmode='linear', tick0=0, dtick=0.75, tickangle = degrees), font=dict(family="Arial", size=20))
ucm.show()
ucm.write_html("..\\reports\\html\\unempBars.html")
ucm.write_image("..\\reports\\png\\unempBars.png", width=1280, height=960)
```
